```python
#preliminary housekeeping
home_dir = '/data/home/iballard/fd/'
subj_file = home_dir + 'subjects.txt'
sub_list = list(np.loadtxt(subj_file,'string'))
os.chdir(home_dir)
exp_list = ['ser_8mm','sim_8mm','loc']
```

```python
bad_mov = {'sub':[],'exp':[],'run':[],'disp':[],'num_spikes':[]}
thresh = 1
for exp in exp_list:
    for sub in sub_list:
        for run in range(1,4):
            
            sub_path = sub_path = home_dir + 'analysis/' + exp + '/' + sub + '/preproc/run_'\
                + str(run)
            realign = sub_path + '/realignment_params.csv'
            if os.path.exists(realign):
                realign = pd.read_csv(realign)
                
                max_disp = max(realign['displace_rel'].values)
                num_spikes = np.sum(realign['displace_rel'].values > thresh)
                if max_disp > thresh:
                    bad_mov['sub'].append(sub)
                    bad_mov['exp'].append(exp)
                    bad_mov['run'].append(run)
                    bad_mov['disp'].append(max_disp)
                    bad_mov['num_spikes'].append(num_spikes)
bad_mov = pd.DataFrame(bad_mov)
bad_mov.sort(['sub'])
```
